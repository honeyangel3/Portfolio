Instruction: Perform a forward and backward propagation in python using the inputs from Laboratory Task 2

```python
import numpy as np
```

```python
x = np.array([1, 0, 1])          # input vector
y = np.array([1])                # target output
lr = 0.001  
```

```python
# Hidden Unit Weights
W_hidden = np.array([
    [0.2, -0.3],
    [0.4,  0.1],
    [-0.5, 0.2]
])


b_hidden = np.array([-0.4, 0.2])
```

```python
# Output layer weights and bias
W_output = np.array([-0.3, -0.2])
b_output = 0.1
```

```python
# Activation Functions
def relu(z):
    return np.maximum(0, z)

def relu_derivative(z):
    return (z > 0).astype(float)
```

## Forward Pass

```python
Z_hidden = np.dot(x, W_hidden) + b_hidden
A_hidden = relu(Z_hidden)

Z_output = np.dot(A_hidden, W_output) + b_output
A_output = relu(Z_output)
```

```python
# Compute error
error = y - A_output
loss = 0.5 * (error ** 2)
```

```python
print("FORWARD PASS:")
print("Z_hidden:", Z_hidden)
print("A_hidden (ReLU):", A_hidden)
print("Z_output:", Z_output)
print("A_output (y_hat):", A_output)
print("Error:", error)
print("Loss:", loss)
```

## Back Propagation

```python
delta_output = error * relu_derivative(Z_output)

# Hidden layer delta
delta_hidden = (delta_output * W_output) * relu_derivative(Z_hidden)
```

```python
# Gradients
grad_W_output = A_hidden * delta_output
grad_b_output = delta_output

grad_W_hidden = np.outer(x, delta_hidden)
grad_b_hidden = delta_hidden
```

```python
# Weight Updates
W_output += lr * grad_W_output
b_output += lr * grad_b_output

W_hidden += lr * grad_W_hidden
b_hidden += lr * grad_b_hidden
```

## Results

```python
print("\nBACKWARD PASS:\n")
print("delta_output:", delta_output)
print("delta_hidden:", delta_hidden)
print("grad_W_output:", grad_W_output)
print("grad_W_hidden:\n", grad_W_hidden)
print("_________________________")
print("\nUPDATED PARAMETERS:\n")
print("Updated W_output:", W_output)
print("Updated b_output:", b_output)
print("Updated W_hidden:\n", W_hidden)
print("Updated b_hidden:", b_hidden)
```
