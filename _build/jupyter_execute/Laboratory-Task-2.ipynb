### Laboratory Task 2

Instruction: Perform a single forward pass and compute for the error.

```python
import numpy as np
```

```python
# Given Data
x = np.array([1, 0, 1])      
y = np.array([1]) 
```

```python
# Hidden Unit Weights
weights_hidden = np.array([
    [0.2, -0.3], 
    [0.4,  0.1],   
    [-0.5, 0.2]  
])

theta_hidden = np.array([-0.4, 0.2])
```

```python
# Output layer weights and bias
weights_output = np.array([-0.3, -0.2])
theta_output = 0.1
```

```python
# Activation function (ReLU)
def relu(z):
    return np.maximum(0, z)
```

```python
# Forward pass
Z_hidden = np.dot(x, weights_hidden) + theta_hidden
H = relu(Z_hidden)

Z_output = np.dot(H, weights_output) + theta_output
y_hat = relu(Z_output)
```

```python
# Compute error
squared_error = (y - y_hat) ** 2
half_squared_error = 0.5 * squared_error
```

```python
# Display results
print("Z_hidden:", Z_hidden)
print("H (after ReLU):", H)
print("Z_output:", Z_output)
print("Predicted y_hat:", y_hat)
print("Squared Error:", squared_error)
print("Half Squared Error:", half_squared_error)
```
